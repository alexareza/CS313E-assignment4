```python
import math
import numpy as np
import matplotlib.pyplot as plt

xsizes = [5, 15, 50]
for msize in xsizes:
    t = np.arange(0, msize, 0.1)

    plt.plot(t, 2 ** 10 * t + 2 ** 10, 'red', t, t ** 3.5 - 1000, 'blue',
             100 * t ** 2.1 + 50, 'green')

    plt.xlim(0, msize)
    plt.grid()
    plt.rcParams["figure.figsize"] = (7, 7)
    plt.savefig(f'graph_size{msize}.png', dpi=400)
    plt.show()
```
